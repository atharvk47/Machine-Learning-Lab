# Splitting of Dataset

<h2> Importing Libraries </h2>

```python
import pandas as pd
import numpy as np
```

<h2> Importing Dataset </h2>

```python
df = pd.read_excel("Salary_Dataset_02.xls") # Reading excel file
```

```python
df.head()
```

<h2> Dividing into Target and Predictor Variables </h2>

```python
# In X we store predictor variables.
# Here X will be dataframe because it has more than one features(columns).
X = df.iloc[:,:-1]
#In y we store target variables.
#Here y will be pandas series as it has only one feature(column).
y = df.iloc[:,-1]

```

<h2> Divide Data as Training Set and Testing Set </h2>

```python
from sklearn.model_selection import train_test_split   #importing from scikit-learn an advanced library
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state = 90)
#For all random datasets, each assign with a random_state value. It means one random_state value has a fixed dataset.
#It means every time we run code with random_state value 1, it will produce the same splitting datasets.

```
